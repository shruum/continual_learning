```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
fold = '/home/mbosc/srv-nas/mammoth_icpr/perturbation/' # <YOUR PATH HERE>
listo = list(filter(lambda x: x.endswith('.npy') and x.startswith('loss') and '2021' in x, os.listdir(fold)))
```

```python
for j, x in enumerate(listo):
    data = np.load(fold+x).mean(0).mean(1) # first mean on experiment number (in case you have more than 1) and 
                                           # second mean across tasks
    plt.plot(pd.Series(data).rolling(5).mean(), marker='o')
plt.legend([x.split('loss_')[1].split('_')[0] for x in listo])
plt.gca().set_ylabel("Train loss")
plt.gca().set_xlabel("sigma")
```
